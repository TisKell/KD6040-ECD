```python
# Import Libraries

import matplotlib.pyplot as plt  # For plotting
import numpy as np  # For numerical operations and trendlines
from scipy.stats import linregress  # For linear trendlines 
import pandas as pd # For reading excel files
import csv #for reading txt files
```

```python
# Specify the full or relative file path
Lifile_pathColour = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water- Li\LiClO4 in DI 20mVs Colour.txt'  
Lifile_pathBleached = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water- Li\LiClO4 in DI 20mVs Bleached.txt'  
Znfile_pathColour = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water- Zn\Zinc - Water 221124.txt'  
Znfile_pathBleached = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water- Zn\Zinc - Water - Clear 221124 Saving issues.txt'  
Alfile_pathColour = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water - Al\AlCl3 DI Blue.txt'  
Alfile_pathBleached = r'C:\Users\w19004417\OneDrive - Northumbria University - Production Azure AD\Level 6\KD6040 Diss\Python\Water\Transmission\Water - Al\AlCl3 DI Bleached Attempt 2.txt'  


# Function to read data from a file and return x and y lists
def read_data(file_path):
    x_values = []
    y_values = []
    
    with open(file_path, mode='r') as file:
        reader = csv.DictReader(file)

         # Strip spaces from headers
        reader.fieldnames = [field.strip() for field in reader.fieldnames]
        
        for row in reader:
            x_values.append(float(row['Wavelength nm.']))
            y_values.append(float(row['T%']))
            
    
        
    return x_values, y_values

# Read data from all files
x1, y1 = read_data(Lifile_pathColour)
x2, y2 = read_data(Lifile_pathBleached)
x3, y3 = read_data(Znfile_pathColour)
x4, y4 = read_data(Znfile_pathBleached)
x5, y5 = read_data(Alfile_pathColour)
x6, y6 = read_data(Alfile_pathBleached)

# Plot the data from both files on the same graph
plt.figure(figsize=(15, 6))
plt.plot(x1, y1, label='Lithium Blue Colour', color='blue')  # Data from the first file
plt.plot(x2, y2, label='Lithium Bleached', color='blue', linestyle = 'dashed')   # Data from the second file
plt.plot(x3, y3, label='Zinc Blue Colour', color='red')  # Data from the first file
plt.plot(x4, y4, label='Zinc Bleached', color='red', linestyle = 'dashed')   # Data from the second file
plt.plot(x5, y5, label='Aluminium Blue Colour', color='black')  # Data from the first file
plt.plot(x6, y6, label='Aluminium Bleached', color='black', linestyle = 'dashed')   # Data from the second file


# Add labels and title
plt.xlabel('Wavelength (nm)',fontsize=12, fontname="Times New Roman", fontweight="medium")
plt.ylabel('T%',fontsize=12, fontname="Times New Roman", fontweight="medium")
plt.title('UV-Vis Scans for LiClO4, AlCl3 and ZnCl2 in DI, in Coloured States',fontsize=20, fontname="Times New Roman", fontweight="bold")
plt.legend()

# Show the plot
plt.show()
#plt.savefig("LiClO4 in DI UV-Vis Scans.png")
```
